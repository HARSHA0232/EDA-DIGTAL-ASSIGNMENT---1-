```
#import libraries
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import numpy as np
import pandas as pd
```

```
url = '/content/student-mat.csv'
data = pd.read_csv(url)

# Selecting features and target variable based on your dataset
X = data[['age', 'studytime', 'failures', 'freetime', 'goout']]  # Example features (select relevant ones based on your dataset)
y = data['G3']  # Target variable (G3: final grade)
```

```
# Split the dataset into training and testing sets (70% training, 30% testing)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Check the shapes of the resulting sets to ensure everything is correct
print(f"Training set size: {X_train.shape}, Testing set size: {X_test.shape}")
```

```
# Initialize and train the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)
```

```
# Make predictions on the test set
y_pred = model.predict(X_test)
```

```
# Calculate Mean Squared Error
mse = mean_squared_error(y_test, y_pred)

# Calculate R-squared (R²)
r2 = r2_score(y_test, y_pred)

# Calculate Mean Absolute Error
mae = mean_absolute_error(y_test, y_pred)

# Calculate Root Mean Squared Error
rmse = np.sqrt(mse)

# Calculate Accuracy (within 0.5 tolerance)
accuracy = np.mean(np.abs(y_pred - y_test) <= 0.5)

# Display the results
print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared (R²): {r2}")
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"Accuracy (within 0.5 tolerance): {accuracy}")
```
