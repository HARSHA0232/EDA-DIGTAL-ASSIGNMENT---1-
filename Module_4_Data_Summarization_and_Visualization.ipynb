### 1. Statistical Summary Measures

```python

# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load the dataset
url = 'https://github.com/HARSHA0232/EDA-DIGTAL-ASSIGNMENT---1-/blob/main/student-mat.csv?raw=true'
data = pd.read_csv(url)

# Displaying the statistical summary of the dataset
summary = data.describe()
summary
```

### 2. Data Elaboration

```python

# Select only numeric columns
numeric_data = data.select_dtypes(include=[np.number])

# Calculate additional statistical measures
data_summary = pd.DataFrame({
    'Mean': numeric_data.mean(),
    'Median': numeric_data.median(),
    'Variance': numeric_data.var(),
    'Standard Deviation': numeric_data.std(),
    'Skewness': numeric_data.skew(),
    'Kurtosis': numeric_data.kurt()
})

# Display the detailed summary for numeric columns only
print(data_summary)
```

### 3. 1-D Statistical Data Analysis

```python


# 1-D Analysis: Distribution plots for 'G1' (First Term Grade)
plt.figure(figsize=(10, 6))
sns.histplot(data['G1'], bins=20, kde=True, color='red')
plt.title('Distribution of First Term Grades (G1)')
plt.xlabel('First Term Grade (G1)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

```

### 4. 2-D Statistical Data Analysis

```python


# 2-D Analysis: Scatter plot for 'G1' vs 'G3'
plt.figure(figsize=(8, 5))
plt.scatter(data['G1'], data['G3'], color='red')
plt.title('Scatter Plot: First Term Grade (G1) vs Final Grade (G3)')
plt.xlabel('First Term Grade (G1)')
plt.ylabel('Final Grade (G3)')
plt.grid(True)
plt.show()
```

### 5. n-D Statistical Data Analysis

```python


# n-D Analysis: Pair plot for multiple variables
sns.pairplot(data[['G1', 'G2', 'G3', 'absences']], diag_kind='kde')
plt.suptitle('Pair Plot of Grades and Absences', y=1.02)
plt.show()
```

### 6. Contingency Tables

```python


# Creating a contingency table for 'address' and 'Pstatus'
contingency_table = pd.crosstab(data['address'], data['Pstatus'])
contingency_table

```

### 7. Visualization: Scatter Plots, Dot Charts, and Bar Plots

```python


# Scatter plot for 'absences' vs 'G3' (Final Grade)
plt.figure(figsize=(8, 5))
sns.scatterplot(x='absences', y='G3', data=data, color='red')
plt.title('Scatter Plot: Absences vs Final Grade (G3)')
plt.xlabel('Number of Absences')
plt.ylabel('Final Grade (G3)')
plt.grid(True)
plt.show()
```

```python


# Dot chart for 'Fedu' (Father's Education) vs 'G3' (Final Grade)
plt.figure(figsize=(8, 5))
plt.plot(data['Fedu'], data['G3'], 'o', color='red')
plt.title("Dot Chart: Father's Education (Fedu) vs Final Grade (G3)")
plt.xlabel("Father's Education Level (Fedu)")
plt.ylabel('Final Grade (G3)')
plt.grid(True)
plt.show()
```

```python

# Bar plot for 'Fedu' (Father's Education) vs 'G3' (Final Grade)
plt.figure(figsize=(10, 6))
sns.barplot(x='Fedu', y='G3', data=data, palette='viridis')
plt.title("Bar Plot: Father's Education (Fedu) vs Final Grade (G3)")
plt.xlabel("Father's Education Level (Fedu)")
plt.ylabel('Final Grade (G3)')
plt.grid(True)
plt.show()

```

### 3. Handling Categorical Data: Encoding and Visualization

```python

# Creating a copy of the original data for encoding
encoded_data = data.copy()

# Label encoding for categorical columns 'sex' and 'paid'
encoded_data['sex'] = encoded_data['sex'].map({'male': 0, 'female': 1})  # Assuming 'sex' column has 'male' and 'female'
encoded_data['paid'] = encoded_data['paid'].map({'yes': 1, 'no': 0})    # Assuming 'paid' column has 'yes' and 'no'

# Displaying the encoded data
encoded_data.head()

```

### 4. Statistical Analysis with Encoded Data

```python

# Select only numeric columns for calculating statistical measures
numeric_encoded_data = encoded_data.select_dtypes(include=[np.number])

# Calculating statistical measures with encoded categorical data
encoded_data_summary = pd.DataFrame({
    'Mean': numeric_encoded_data.mean(),
    'Median': numeric_encoded_data.median(),
    'Variance': numeric_encoded_data.var(),
    'Standard Deviation': numeric_encoded_data.std(),
    'Skewness': numeric_encoded_data.skew(),
    'Kurtosis': numeric_encoded_data.kurt()
})

# Displaying the detailed summary for the dataset with encoded categorical variables
encoded_data_summary
```

### 5. Updated Visualizations with Encoded Data

```python

# Assuming the encoded_data DataFrame is already created

# Pair plot with selected columns from the encoded data
sns.pairplot(encoded_data[['G1', 'G2', 'G3', 'studytime', 'failures', 'paid']], hue='paid')
plt.suptitle('Pair Plot with Encoded Categorical Variables', y=1.02)
plt.show()
```

```python
# Selecting relevant columns for the heatmap
heatmap_columns = encoded_data[['G1', 'G2', 'G3', 'studytime', 'failures', 'paid', 'sex']]

# Heatmap with the selected columns
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_columns.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap with Encoded Data')
plt.show()
```
